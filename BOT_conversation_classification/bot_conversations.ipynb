## <center>Feature Engineering<center>

```python
df.head()
```

```python
# Are there only single digit numbers in the o/p (not all zeros)
df['all_single_digits'] = ((df[num_cols]==0).sum(axis=1)!=10) & ((df[num_cols]<10).sum(axis=1) == 10)
df['all_single_digits'] = df['all_single_digits'].astype(int)

# Are 10 numbers in a non-decreasing order?
df['is_non_decreasing'] = (df['num2']>=df['num1']) & (df['num3']>=df['num2']) & (df['num4']>=df['num3']) & (df['num5']>=df['num4']) & (df['num6']>=df['num5']) & (df['num7']>=df['num6']) & (df['num8']>=df['num7']) & (df['num9']>=df['num8']) & (df['num10']>=df['num9'])
df['is_non_decreasing'] = df['is_non_decreasing'].astype(int)

# Are the following 9 numbers divisible by the first number?
df['is_div_by_num1'] = (df['num2']%df['num1']) + (df['num3']%df['num1']) + (df['num4']%df['num1']) + (df['num5']%df['num1']) + (df['num6']%df['num1']) + (df['num7']%df['num1']) + (df['num8']%df['num1']) + (df['num9']%df['num1']) + (df['num10']%df['num1']) 
df['is_div_by_num1'] = (df['is_div_by_num1']==0).astype(int)

# Is the next number of o/p divisible by preceeding number?
df['5_times_prev'] = (df['num1']==0) | ((df['num2']/df['num1']==5) & (df['num3']/df['num2']==5) & (df['num4']/df['num3']==5) & (df['num5']/df['num4']==5) & (df['num6']/df['num5']==5) & (df['num7']/df['num6']==5) & (df['num8']/df['num7']==5) & (df['num9']/df['num8']==5) & (df['num10']/df['num9']==5) )
df['5_times_prev'] = df['5_times_prev'].astype(int)

# Is the difference between consecutive numbers less than 10?
df['is_diff_less_than_10'] = ((df['num2']-df['num1'])<10) & ((df['num3']-df['num2'])<10) & ((df['num4']-df['num3'])<10) & ((df['num5']-df['num4'])<10) & ((df['num6']-df['num5'])<10) & ((df['num7']-df['num6'])<10) & ((df['num8']-df['num7'])<10) & ((df['num9']-df['num8'])<10) & ((df['num10']-df['num9'])<10)
df['is_diff_less_than_10'] = df['is_diff_less_than_10'].astype(int)

# Is the next number greater than previous number by exactly 5?
df['is_diff_5'] = ((df['num2']-df['num1'])==5) & ((df['num3']-df['num2'])==5) & ((df['num4']-df['num3'])==5) & ((df['num5']-df['num4'])==5) & ((df['num6']-df['num5'])==5) & ((df['num7']-df['num6'])==5) & ((df['num8']-df['num7'])==5) & ((df['num9']-df['num8'])==5) & ((df['num10']-df['num9'])==5)
df['is_diff_5'] = df['is_diff_5'].astype(int)

# Is the difference between consecutive numbers less than 10 but not all differences are equal to 5?
df['diff_less_than_10_but_not_5'] = df['is_diff_less_than_10'] * (1-df['is_diff_5'])
df['diff_less_than_10_but_not_5'] = df['diff_less_than_10_but_not_5'].astype(int)

df.head(16)
```

## <center>Modelling<center>

### Basic Classification Models

```python
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X = df.iloc[:, 11:].values
y = df.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
# Classifiers
clf1 = RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=0)
clf2 = LogisticRegression()
clf3 = SVC(kernel='rbf', random_state=0)
```

```python
# Fitting data to the models
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)

# Predicting
y_pred_1 = clf1.predict(X_test)
y_pred_2 = clf2.predict(X_test)
y_pred_3 = clf3.predict(X_test)
```

### Results

```python
cm1 = confusion_matrix(y_test, y_pred_1)
cm2 = confusion_matrix(y_test, y_pred_2)
cm3 = confusion_matrix(y_test, y_pred_3)

print('\nRandom Forest Model,    Accuracy: ',accuracy_score(y_test, y_pred_1),'\n', cm1)
print('\nLogistic Regression Model,    Accuracy: ',accuracy_score(y_test, y_pred_2),'\n', cm2)
print('\nSupport Vector Classifier Model,    Accuracy: ',accuracy_score(y_test, y_pred_3),'\n', cm3)
```

### Neural Network Model

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
y_train_nn = onehotencoder.fit_transform(y_train.reshape(-1,1)).toarray()
y_test_nn = onehotencoder.transform(y_test.reshape(-1,1)).toarray()

import keras
from keras.models import Sequential
from keras.layers import Dense
    
# Building the Neural Network
model = Sequential()
    
model.add(Dense(units=5, kernel_initializer='glorot_uniform', activation='relu', input_dim=7))
model.add(Dense(units=5, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train_nn, batch_size=32, nb_epoch=20)

# Predictions
y_pred_nn = model.predict(X_test)
```

```python
y_pred_4 = np.argmax(y_pred_nn, axis=1)
cm4 = confusion_matrix(y_test, y_pred_4)
print('\nNeural Network Model,    Accuracy: ',accuracy_score(y_test, y_pred_4),'\n', cm4)
```
