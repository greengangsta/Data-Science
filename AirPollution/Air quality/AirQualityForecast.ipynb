```python
data = pd.read_excel('AirQualityUCI.xlsx',index_col = 'Date',parse_dates = True)
```

```python
data['Day'] = pd.DatetimeIndex(data['Date']).day
data['month'] = pd.DatetimeIndex(data['Date']).month
data['year'] = pd.DatetimeIndex(data['Date']).year
```

```python
data = data.drop(columns = ['Date','Time'])
data.columns
```

```python
plt.figure(figsize= (15,10))

plt.subplot(311)
data["CO(GT)"].plot(kind = 'line')

plt.subplot(312)
data["NOx(GT)"].plot(kind = 'line')

plt.subplot(313)
data["C6H6(GT)"].plot(kind = 'line')
```
