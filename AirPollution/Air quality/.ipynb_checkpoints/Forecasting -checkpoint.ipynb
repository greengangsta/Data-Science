```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('AirQualityUCI.xlsx',parse_dates = [['Date','Time']])
```

```python
data.columns
```

```python
data = data.set_index('Date_Time')
data.head(2)
```

```python
data.loc[data['CO(GT)'] == -200, 'CO(GT)'] = 2.5
co_data = pd.DataFrame(data['CO(GT)'],columns = ['CO(GT)'])
co_data.head()
```

```python
co_data.info()
```

```python
train_data = co_data.iloc[:8001]
test_data = co_data.iloc[8000:]
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
fitted_model = ExponentialSmoothing(train_data['CO(GT)'],trend = 'add',seasonal = 'add', seasonal_periods = 240).fit()
```

```python
test_predictions = fitted_model.forecast(1357)
```

```python
co_data.plot(figsize = (15,5))
```

```python
train_data['CO(GT)'].plot(legend = True,label = 'TRAIN', figsize= (15,5))
test_data['CO(GT)'].plot(legend = True,label = 'TEST')
```

```python
train_data['CO(GT)'].plot(legend = True,label = 'TRAIN', figsize= (15,5))
test_data['CO(GT)'].plot(legend = True,label = 'TEST')
test_predictions.plot(legend = True,label = 'Test_predictions')
```
